```python
from telethon.sync import TelegramClient 
import datetime
import pandas as pd
import asyncio
import sqlite3
import telethon
```

```python
name = 'anon' 
api_id = 1047847
api_hash = '688e90800dfdcbe42c16c3bd0b173752'
chat = ['africademics', 't.me/joinchat/TqvEknRFT6oMRfCb']
```

```python
async def scholarship_list(chat, name='anon', api_id=1047847, api_harsh='688e90800dfdcbe42c16c3bd0b173752'):
    from datetime import datetime, date
    date = date.today()
    time = datetime.now()
    conn = sqlite3.connect('C:/Users/morara/databases/scholarships.db')
    for grp in chat:
        mesgs = [] # stores all our data in the format SENDER_ID, MSG

        async with TelegramClient(name, api_id, api_hash) as client:
             async for message in client.iter_messages(grp):
                mesgs.append([message.sender_id, message.text])


        df = pd.DataFrame(mesgs, columns=['SENDER', 'MESSAGE']) # creates a new dataframe
        df = df.dropna()
        data_sc = df['MESSAGE'].head(500).tolist()

        data_sc_msg = [time]
        Finance_msg = [time]
        
        for i in range(len(data_sc)):
            msg = data_sc[i]
            if 'data' in msg:
                data_sc_msg.append(msg)
            elif 'finance' in msg:
                Finance_msg.append(msg)
                
        Finance_msg_df = pd.DataFrame(Finance_msg)
        data_sc_df = pd.DataFrame(data_sc_msg)
        Finance_msg_df['Schol_finace'] = Finance_msg_df[0]
        data_sc_df['Schol_d_science'] = data_sc_df[0]
        
        data_sc_df = data_sc_df.drop(columns=0)
        Finance_msg_df = Finance_msg_df.drop(columns=0)
        
        data_sc_df.to_csv(f'D:/scholarships/data_sc_scholarships_{date}_.csv', encoding='utf-8')
        Finance_msg_df.to_csv(f'D:/scholarships/finance_scholarships_{date}_.csv', encoding='utf-8')
        data_sc_df.to_sql('Datascience', conn, if_exists='replace', index=False)
        Finance_msg_df.to_sql('finance', conn, if_exists='replace', index=False)
        fin_sc = pd.read_sql('SELECT * FROM finance', conn)
        data_sc_ = pd.read_sql('SELECT * FROM Datascience', conn)
        return fin_sc, data_sc_
        break
```

```python
await scholarship_list(chat)
```
